```python
from skimage import data, img_as_ubyte
from skimage.feature import hog
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
```

**Load image from Scikit-Image dataset**

```python
image = data.astronaut()
```

**Show image with matplotlib**

```python
plt.imshow(image)
```

**Convert image to grayscale**

```python
gray_img = rgb2gray(image)
```

**Show grayscale image**

```python
plt.imshow(gray_img, cmap='gray')
```

**Extract HOG feature from image**

Result contains feature vector and hog image. 

```python
feature, hog_image = hog(gray_img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(2, 2), visualize=True, block_norm='L2')
```

**Plot feature vector with bar plot to show histogram**

```python
plt.bar(list(range(len(feature))), feature)
```

**Checking feature vector dimension**

```python
feature.shape
```

**Visualize HOG feature**

```python
plt.imshow(hog_image, cmap='gray')
```

**Focus on specific cell**

```python
plt.imshow(hog_image[0:16,0:16], cmap='gray')
```
